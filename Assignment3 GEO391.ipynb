Clip precipitation and temperature data in a state you chose using shapefile.
And then, you save raster data as tif file from nc file. Tiff and nc file are a kind of raster data.

```python
import fiona
import rasterio as rio
import rasterio.mask
from rasterio.plot import show
from rasterio.warp import calculate_default_transform, reproject, Resampling

# read shape file
with fiona.open("./State_MT.shp", "r") as shapefile:
    shape = [feature["geometry"] for feature in shapefile]

# read imagery file and clip it with shape file
with rasterio.open('./precip_F2015-09-01_T2016-08-31.nc') as src:
    out_image, out_transform = rasterio.mask.mask(src, shape, crop=True)
    out_meta = src.meta

with rasterio.open('./temp_F2015-09-01_T2016-08-31.nc') as src2:
    out_image2, out_transform2 = rasterio.mask.mask(src2, shape, crop=True)
    out_meta2 = src2.meta

#Save clipped imagery
out_meta.update({"driver": "GTiff",
                 "height": out_image.shape[1],
                 "width": out_image.shape[2],
                 "transform": out_transform})

#Change the file name every time you save it using the below code if you want to run the code.  
with rasterio.open("./montana_pre.tif", "w", **out_meta) as dest1:
    dest1.write(out_image)  

out_meta2.update({"driver": "GTiff",
                 "height": out_image2.shape[1],
                 "width": out_image2.shape[2],
                 "transform": out_transform2})

with rasterio.open("./montana_temp.tif", "w", **out_meta2) as dest2:
    dest2.write(out_image2)
```

Calculate mean precipitation and temperature (See the material you have)

```python
#To calculate mean precipitation 
rPrecip=rio.open('./montana_pre.tif')
npPrecip=rPrecip.read(masked=True)
npPrecip=(npPrecip.T*rPrecip.scales).T
npmeanPrecip=np.mean(npPrecip, axis=0)

#To calculate mean temperature
rtemp=rio.open('./montana_temp.tif')
nptemp=rtemp.read(masked=True)
nptemp=(nptemp.T*rtemp.scales).T
meantemp=np.mean(nptemp, axis=0)
meantemp=(meantemp/10 - 32)*5/9  # To convert F to C degree
```

The datetime module supplies classes for manipulating dates and times. You should make a datetime during 2019-09-01 - 2020-08-31 referencing the below coding.

```python
from datetime import datetime as dt
from datetime import timedelta

#Setting the start and end date
strdt = dt.strptime("2015-09-01", '%Y-%m-%d')  # Start
enddt = dt.strptime("2016-08-31", '%Y-%m-%d')  # End

# Calculate the range of date you focus on
days_num = (enddt - strdt).days +1 
datelist = []
for i in range(days_num):
    datelist.append(strdt + timedelta(days=i))


# Check the datelist you made
for d in datelist:
    print(d.strftime("%Y-%m-%d"))
```

Extract mean value on each date using for loop

```python
pre=[]
for i in range(366):
    data=np.mean(npPrecip[i,:,:])
    pre.append(data)

temp=[]
for i in range(366):
    data1=np.mean(nptemp[i,:,:])
    temp.append(data1)  
```

Make the array of precipitation and temperature using np.array. After that, temperature should be converted from F to C defree

```python
pre=np.array(pre)
temp=np.array(temp)
temp=(temp/10 - 32)*5/9
```
