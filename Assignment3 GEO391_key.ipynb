```python
import numpy as np
import matplotlib.pyplot as plt
import rasterio as rio
rtemp = rio.open('./agg_met_tmmx_1979_CurrentYear_CONUS.nc.nc4')
```

```python
rtemp.meta
```

```python
import geopandas as gpd
gp_counties = gpd.read_file('./USA_1990-2000_Population_Change.shp')
gp_montana = gp_counties[gp_counties['NAME'] == 'Montana']
```

```python
gp_montana
```

```python
import rasterio.mask
npMontana, aff = rasterio.mask.mask(rtemp, gp_montana['geometry'],filled=False)
```

Temperature unit is K. In order to convert to the unit correctly, you need to multiple it by scale factor and then need to add offset value (220). 

```python
npMontana=(npMontana.T*rtemp.scales).T
meantemp=np.mean(npMontana, axis=0)
meantemp=meantemp+220
```

## Answer Q3

```python
from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib import cm
from rasterio.plot import show
fig = plt.figure(figsize=(5,3))
ax = fig.add_subplot()
spatial_extent = rasterio.plot.plotting_extent(rtemp) 
raster_image=plt.imshow(meantemp, extent = spatial_extent)
idx = gp_counties['NAME'].isin(['Alaska','Puerto Rico','Hawaii'])
gp_counties = gp_counties[~idx]
vector_image = gp_counties.plot(ax=ax,facecolor="none",edgecolor='black', lw=1)

ax.set_title('USA')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
cb = fig.colorbar(raster_image, orientation='vertical')
cb.set_label('Temperature (K)', fontsize= 14)
fig.tight_layout()
plt.savefig('q3key_new.png',format="png", dpi=300)
```

## If you want to add the correct date label on the x axis, use datatime module.
The datetime module supplies classes for manipulating dates and times. You should make a datetime during 2019-09-01 - 2020-08-31 referencing the below coding.

```python
from datetime import datetime as dt
from datetime import timedelta

#Setting the start and end date
strdt = dt.strptime("2019-09-01", '%Y-%m-%d')  # Start
enddt = dt.strptime("2020-08-31", '%Y-%m-%d')  # End

# Calculate the range of date you focus on
days_num = (enddt - strdt).days +1 
datelist = []
for i in range(days_num):
    datelist.append(strdt + timedelta(days=i))


# Check the datelist you made
for d in datelist:
    print(d.strftime("%Y-%m-%d"))
```

Make the plot again

```python
#fig, ax1 = plt.subplots()
plt.rcParams['figure.figsize'] = (7, 4)
fig = plt.figure()
ax1 = fig.add_subplot()
ax1.plot(datelist,(220+npMontana.mean(axis=(1,2))-273.15)) # you can convert K into Degree C here
ax1.set_ylabel('Temperature (degree C)', color='b',fontsize=15)
ax1.set_xlabel('Time', color='k',fontsize=15)
ax1.tick_params('y', colors='b',labelsize='15')
```
